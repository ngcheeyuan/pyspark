```
%sql
SELECT corr(age,crew)
      ,corr(Tonnage,crew)
      ,corr(passengers,crew)
      ,corr(length,crew)
      ,corr(cabins,crew)
      ,corr(passenger_density,crew)
FROM cruise_csv
```

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('cruise').getOrCreate()
```

```
from pyspark.ml.feature import StringIndexer
```

```
indexer = StringIndexer(inputCol = 'Cruise_line',
                       outputCol = 'cruise_cat')
indexed = indexer.fit(df).transform(df)
indexed.head(3)
```

```
indexed.show(5)
```

```
from pyspark.ml.regression import LinearRegression
```

```
df.printSchema()
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```
indexed.columns
```

```
assembler = VectorAssembler(inputCols = [ 'Age',
 'Tonnage',
 'passengers',
 'length',
 'cabins',
 'passenger_density',
 'cruise_cat'],
                           outputCol='features')
```

```
output = assembler.transform(indexed)
```

```
output.head(1)
```

```
final_data = output.select('features','crew')
```

```
final_data.show()
```

```
train_data , test_data = final_data.randomSplit([0.7,0.3])
```

```
train_data.describe().show()
```

```
lr = LinearRegression(labelCol = 'crew')
```

```
lr_model = lr.fit(train_data)
```

```
train_results = lr_model.evaluate(train_data)
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

residuals = [i[0] for i in train_results.residuals.collect()]
predictions = [i[1] for i in train_results.predictions.collect()]

sns.scatterplot(x = predictions,
                y = residuals)
```

```
lr_results = lr_model.evaluate(test_data)
```
